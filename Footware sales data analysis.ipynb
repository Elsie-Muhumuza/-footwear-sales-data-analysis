This dataset provides sales transaction information for a variety of products across different brands, sizes, and dealers for the month of January 2023. Each row represents a single transaction and includes the following columns:



1.   Date: The date of the transaction.
2.   Product: The type of product sold (e.g., Sneakers, Flats, Sandals).

5.   Quantity Sold: The quantity of the product sold in this transaction
6.  Unit Price (₹): The unit price of the product.
7.   Margin (%): The profit margin percentage.
8.  Profit (₹): The profit generated from this transaction.
9.  Net Profit (₹): The net profit obtained after deducting taxes.
10.   Total Revenue (₹): The total revenue generated from this transaction.
11.  Tax (GST %): The Goods and Services Tax (GST) percentage applied.
12.  Tax Amount (₹): The total tax amount paid in this transaction.
13.   Net Tax (₹): The net tax amount after deductions.
14.  Dealer: The dealer responsible for this transaction.
15.   Stock Availability: The availability of stock at the time of the transaction.
16.  Dealer Location: The location of the dealer.

# Data Exploration

Establishing the different data types within the dataset

```python
df.dtypes
```

Cleaning the data, starting with clearing the white spaces in column titles and replacing them with underscores. In addition, replacing Rupee symbol with text for accessibilty

```python
#Replace white space with underscores and symbol with text
df.columns=(df.columns.str.strip().str.replace(" ", "_").str.replace("₹", "rupees"))
df.dtypes
```

Changing column titles to lower case

```python
#Change column names to lower case
df.columns=df.columns.str.lower()
df.dtypes
```

```python
#Removing the underscore that appears before any bracketed word
df.columns=[col.replace("_(", "(")for col in df.columns]
df.dtypes
```

```python
#Removing the underscore that appears before and after "%"
df.columns=[col.replace("_%_", "%")for col in df.columns]
df.dtypes
```

```python
df.head()
```

Converting object types that need to be numerical into floats

```python
#Find columns that contain at least one '%' in any row
percent_cols = [col for col in df.columns if df[col].astype(str).str.contains('%').any()]

#Convert those columns: remove '%', convert to float, divide by 100
for col in percent_cols:
    df[col] = df[col].str.rstrip('%').astype(float) / 100
df.dtypes
```

# Visualization

To determine what brand was generating the most profit, both box and violin plots were used to determine these findings.

On both visualizations, the X-axis represents the categorical variables - the different shoe brands that Big Foot Wholesale has in stock.

The Y-axis represents the numerical variable in form of the Net Profit (Rupees) generated from the month of January 2023.

These models are visualiziing the net profit across the different brands of shoes

### **Boxplot**
A box plot contains the following elements:


*   The box - shows the interquartile range containing the middlie line which represents half the data.
*   Median - divides the box into two, representing the median value
*   Whiskers - lines that exxtend from the box to show the minimum and maximum data range
*   Outliers - individual data beyond the whiskers


The spread of the box represents variabilty for a category i.e a tall box means more variability in the net profit of that particular category. A shorter box means most values of the net profit are clustered tightly.

```python
#Create the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='brand', y='net_profit(rupees)',palette='coolwarm')
plt.title('Big Foot Wholesale - Net Profit against Brand')
plt.xlabel('Brand')
plt.ylabel('Net Profit(rupees)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## **Violin plot**

A violin plot contains elements of a box plot and a kernel density plot.
Its shape represents the density of the data with wider sections indicating higher density and narrower sections representing lower density. The plot is named as such as beacause it borrows its shape from a violin.

The violin plot also has a median - the white dot in the middle of the shape and the thick bar in the centre being the interquartile range

```python
#Create the violinplot
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='brand', y='net_profit(rupees)',palette='coolwarm')
plt.title('Big Foot Wholesale - Net Profit against Brand')
plt.xlabel('Brand')
plt.ylabel('Net Profit(rupees)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### **Comparing Medians across brand categories**
From the above model, Nike presents as the brand generating the highest profit while Liberty has generated the least. This evident from the position of the median lines in their respective plots.

### **Comparing Spread and Outliers**

Nike's spread signifies a wide range in it's profit which could in turn contribute to why it generated the most revenue. In addition, having multiple outliers shows extreme values beyond the expected net profit.
Brands like Nike and Puma have clear outliers on the high side.
Brands like  Liberty and Paragon showcase a low return on net profit with their clustered plots.

The wider boxes for Nike and Woodland indicate greater variability in profit suggesting they both have low & high profit gains.
Outliers for Nike and Puma suggest some unusually large transactions which could be big sales or special events.


### **Violin plot readings**
A wide violin for example Nike, signifies many typical sales at mid profit and some very high-profit deals.

Long tails like Adidas and Woodland show that some deals are bigger, which the business should replicate for similar profit going forward.

Tight and narrow violins seen within the Paragon and Liberty brands signify consistent lower profit

The violin plot shows that net profit for brands like Nike, Woodland, and Reebok not only has higher medians but also greater variability, indicating opportunities for both steady sales and high-value bulk deals. Brands like Paragon and Liberty have tighter, lower profit distributions, suggesting more consistent but lower-value sales. This insight can help the business tailor marketing and sales strategies for each brand
